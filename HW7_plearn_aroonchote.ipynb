## Problem 4

Now we will tackle the challenge of mitigating global climate change using integrated assessment models.  These models try to capture the feedback between human (economic) activity and the climate.  The central premise is that human economic activity results in emissions, which in turn change the climate, and this feedbacks and impacts the economy.  Modeling how human activity results in emissions, how the climate responds to those emissions, then how the climate impacts eco- and human-systems are each complex in their own way, having their own academic disciplines.  To model the interactions between those systems we must take a very simplified approach that often ignores important aspects of the system, to say nothing of the complex natural and societal issues (for instance issues of equity, social justice, etc.).  The art of systems analysis is knowing how your simplifications may impact your conclusions, and carving out a model that is as simple as possible while still addressing the research question.

So what is the research question?  Why take a simple approach to modeling these systems?  What is the place of IAMs?  As discussed by Flannery, IAMs are used to design scenarios or provide other analyses that form the foundation of more detailed asessement (e.g. RCPs, SSPs, SCC).  IAMs can help bound the conditions that other disciplines consider.  In turn more detailed models can highlight the importance of unrepresented feedbacks, which informs the next generation of IAMs in a cycle of model improvement.

Returning to climate mitigation, the fundamental challenge is to reduce emissions to avoid catostrophic ecological impacts with out destroying the global economy.  At the same time, doing nothing will result in ecosystems failures that will also have severe economic impacts.  So the challenge is to act quickly and efficiently to meet or exceed climate targets in a way that minimizes economic impacts. which will inevitably fall hardest on margainalized communities.  We can (and economists do) structure this as a constrained optimization problem using a simplified IAM.

Specifically we will use the Simple Integrated Assessment Model (SIAM) proposed by Dr. Andrew Jarvis of the Univeristy of Lanchaster in the UK.  That model uses six equations, which is a gross over simplification, but forms a marvelous didactic example.  At its heart the model imagines the economy in terms of primary energy consumption over time, $x(t)$.  Historically, $x(t)$ has tracked closely with economic growth (globaly), as described by the gross world product, $z(t)$:

<h4><center>$z(t) = 0.05x^{1.2}$</center></h4>

Each year, most of the energy consumption goes towards running the economy, while some fraction, $\eta$, is invested into infrastructure, $X$, that allows us to extract and utilize more energy.  The relationship between energy consumption, $x$, and capital stock (infrastructure) of invested energy is:
<h4><center>$x=\frac X \lambda$</center></h4>

where $\lambda$ is the residence time of energy flows (i.e. energy storage capacity), assumed to be 1 year.  Infrastructure capital cost degrades at a rate $d$, so that the balance equation for infrastructure is:

<h4><center>$X(t+1)=X(t)+\eta*x(t)-d*X(t)$</center></h4>

So the captial stock of infrastructure next year is what we have this year, less what we loss to degredation, plus what we invest from this year's energy consumption.  In the absence of climate change, the average life span of energy infrastructure is about 40-years, so the degredation rate now is about $d_{0}=1/40=0.025$.  Some work suggests that climate change will have little impact on this degredation rate until global warming, $\Delta T$, exceeds 1 degree celsius, at which point some speculate it will increase linearly:

<h4><center>$d=\max(d_{0};d_{0}+0.004*(\Delta T-1))$</center></h4>

Global warming is a function of the cumulative $CO_{2}$ emissions, $\sum y$ and a simple approximation of the warming associated with cumulative emissions is:

<h4><center>$\Delta T = (\sum{y})/500$</center></h4>

The emissions in any year, $y(t)$, corresponds to the primary energy consumption:

<h4><center>$y(t)=0.06x(t)^{0.8}$</center></h4>

Clearly this model is too simple, but it has many intreresting features that are reflective of the true system.  We will explore some of them in the following problems.  Below is an implementation of the model, with a base-year of 2020.

```python
import numpy as np
def IAM_BAU():
    GWP_0 = 61#127
    PEC_0 = 364#20*np.exp(np.log(GWP_0)/1.2)
    CS_0 = PEC_0
    
    d0 = 0.025
    eta = 0.05
    lamb = 1
    cum_emiss = 397
    
    GWP = np.zeros([81])
    PEC = np.zeros([81])
    CS = np.zeros([81])
    Emissions = np.zeros([81])
    Warming = np.zeros([81])
    Decay = np.zeros([81])
    
    Warming[0] = cum_emiss/400
    Emissions[0] = 0.06*PEC_0**0.8
    Decay[0] = d0
    CS[0] = CS_0+eta*PEC_0-Decay[0]*CS_0
    PEC[0] = CS[0]/lamb
    GWP[0] = 0.05*PEC[0]**1.2
    
    for i in range(1,81):
        Warming[i] = (np.sum(Emissions[0:i])+cum_emiss)/400
        Emissions[i] = 0.06*PEC[i-1]**0.8
        Decay[i] = max(d0,d0+0.004*(Warming[i]-1))
        CS[i] = CS[i-1]+eta*PEC[i-1]-Decay[i]*CS[i-1]
        PEC[i] = CS[i]/lamb
        GWP[i] = 0.05*PEC[i]**1.2
    return GWP,Warming,CS,Emissions,PEC

orig_GWP,orig_Warming,CS,Emissions,PEC = IAM_BAU()
```

## Part D
Clearly some mitigative effort is required.  One way to represent this in the SIAM model is to imagine two parallel economies: a Low Carbon Economy (LCE), which contributes to energy production and global capital stock but does not make emissions, and a High Carbon Economy (HCE) which is roughly the equivalent of our previous example.  Every year some investment, $i$, can be made which invests productive energy from the HCE to increasing capital stock in the LCE.  This could be something akin to a carbon tax, or perhaps some governmental intervention like the Green New Deal.

Now, rather than one captial stock flow equation we have two, one for the LCE:
<h4><center>$X_{L}(t+1)=X_{L}(t)+\eta_{L}*x_{L}(t)+i(t)*\eta_{H}*x_{H}(t)-d*X_{L}(t)$</center></h4>

where $\eta_{L}$ and $x_{L}$ are the fraction of LCE primary energy consumption invested in infrustructure expansion and the LCE primary energy consumption, respectively.  $i*\eta_{H}$ is the fraction of HCE energy consumption, $x_{H}$ invested in expanding the LCE infrustructure stock.  $i*\eta_{H}$ represents a transfer of wealth, or investment, from the HCE to the LCE.  The HCE capital stock flow eqaution follows naturally:
<h4><center>$X_{H}(t+1)=X_{H}(t)+(1-\eta_{H})*x_{H}(t)-d*X_{H}(t)$</center></h4>

The IAM_BAU formualtion has been adapted to accomodate this new problem formulation.  In this case, the function inputs are the investments (must range from 0-1) to make in 2020,2040,2060, and 2080.  Investments are assumed to be fixed for 20 years.  Set the boundary, by assuming we transfer all re-investments from the HCE to LCE for the rest of the century.  Plot the Capital Stock break down between HCE and LCE versus time.  If we are investing 100% in LCE from 2020, why do renewables (LCE infrastructure) not take over right away?  Plot the GWP for this extreme solution vs. idealized 3% growth model.  Comment on the results.

```python
def IAM_mit(invest):
    GWP_0 = 61#127
    PEC_LCE_0 = 364*0.13#20*np.exp(np.log(GWP_0)/1.2)
    PEC_HCE_0 = 364-PEC_LCE_0
    CS_LCE_0 = PEC_LCE_0
    CS_HCE_0 = PEC_HCE_0
    
    d0 = 0.025
    eta_HCE = 0.05
    eta_LCE = 0.03
    lamb = 1
    cum_emiss = 397
    
    GWP = np.zeros([81])
    PEC = np.zeros([81])
    CS = np.zeros([81])
    Emissions = np.zeros([81])
    Warming = np.zeros([81])
    Decay = np.zeros([81])
    LCE_CS = np.zeros([81])
    HCE_CS = np.zeros([81])
    PEC_LCE = np.zeros([81])
    PEC_HCE = np.zeros([81])
    
    invests = np.zeros([81])
    invests[0:20] = invest[0]
    invests[20:40] = invest[1]
    invests[40:60] = invest[2]
    invests[60:81] = invest[3]
    
    Warming[0] = cum_emiss/400
    Emissions[0] = 0.06*PEC_HCE_0**0.8
    Decay[0] = d0
    LCE_CS[0] = CS_LCE_0+eta_LCE*PEC_LCE_0+invest[0]*eta_HCE*PEC_HCE_0-d0*CS_LCE_0
    HCE_CS[0] = CS_HCE_0+(1-invest[0])*eta_HCE*PEC_HCE_0-d0*CS_HCE_0
    CS[0] = LCE_CS[0]+HCE_CS[0]
    PEC_LCE[0] = LCE_CS[0]/1
    PEC_HCE[0] = HCE_CS[0]/1
    PEC[0] = PEC_LCE[0]+PEC_HCE[0]
    GWP[0] = 0.05*PEC[0]**1.2
    
    for i in range(1,81):
        Warming[i] = (np.sum(Emissions[0:i])+cum_emiss)/400#OK
        Emissions[i] = 0.06*PEC_HCE[i-1]**0.8#OK
        Decay[i] = max(d0,d0+0.004*(Warming[i]-1))#OK
        LCE_CS[i] = LCE_CS[i-1]+eta_LCE*PEC_LCE[i-1]+invests[i]*eta_HCE*PEC_HCE[i-1]-Decay[i]*LCE_CS[i-1]#OK
        HCE_CS[i] = HCE_CS[i-1]+(1-invests[i])*eta_HCE*PEC_HCE[i-1]-Decay[i]*HCE_CS[i-1]#OK
        CS[i] = LCE_CS[i]+HCE_CS[i]#OK
        PEC_LCE[i] = LCE_CS[i]/1#OK
        PEC_HCE[i] = HCE_CS[i]/1#OK
        PEC[i] = PEC_LCE[i]+PEC_HCE[i] #OK
        GWP[i] = 0.05*PEC[i]**1.2 #OK
    return GWP,Warming,CS,LCE_CS,HCE_CS,Emissions

GWP,Warming,CS,LCE_CS,HCE_CS,Emissions = IAM_mit([1,1,1,1])
```
